```python
import matplotlib.pyplot as plt

# Pie Chart
labels = ["A", "B", "C"]
sizes = [30, 45, 25]

plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Simple Pie Chart")
plt.show()

# Bar Chart
x = ["A", "B", "C"]
y = [10, 20, 15]

plt.bar(x, y)
plt.title("Simple Bar Chart")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    "Category": ["A", "B", "C"],
    "Value": [10, 20, 15]
}

# Bar Plot
sns.barplot(x="Category", y="Value", data=data)
plt.title("Seaborn Bar Plot")
plt.show()

# Line Plot
sns.lineplot(x="Category", y="Value", data=data)
plt.title("Seaborn Line Plot")
plt.show()
```
